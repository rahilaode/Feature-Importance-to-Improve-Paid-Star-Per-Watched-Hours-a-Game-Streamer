# Problem Statement

A leading agency representing content creators on Gaming committed to help streamers increase their revenues by leveraging the rich source of streaming data to provide actionable insights. In this problem, we want to learn whether certain features of streamers will make them more or less likely to receive stars from their audience, in this case PaidStarPerWatchedHour is the dependent variable. 

# Dataset Description

This dataset consists of more than 600 streamers (Indonesia, Vietnam, Philippines). Each row represents data of one unique streamer. There are different sets of facial features extracted using an AI tool (e.g., Self_Esteem_, Character_Facet_, Personal_Values_Facet_, Temperament_), in addition to Gender, Game, Country, Total Follower, and the role they play in the game (e.g., Role_Director).

# 2. Data Preprocessing

```python
df = pd.get_dummies(df)
```

## 2.1. Splitting Data

```python
X = df.drop(['PaidStarPerWatchedHour'], axis = 1)
y = df['PaidStarPerWatchedHour']
```

```python
X
```

## 2.2. Scalling Data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_scaled.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.2,
                                                   random_state = 6)
```

# 3. Select important features from feature_importances_ atributes of model

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
#evaluation of results
def model_evaluation(y_test, y_pred, model_name):
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    results = pd.DataFrame([[model_name, mse, mae, rmse]], 
                       columns = ["Model", "MSE", "MAE", "RMSE"])
    results = results.sort_values(["RMSE"], ascending = False)
    return results
```

## 3.1. Models

```python
#DECISION TREE REGRESSOR
dct_regressor = DecisionTreeRegressor()
dct_regressor.fit(X_train, y_train)
dct_pred = dct_regressor.predict(X_test)

#RANDOM FOREST REGRESSOR
rf_regressor = RandomForestRegressor()
rf_regressor.fit(X_train, y_train)
rf_pred = rf_regressor.predict(X_test)

#KNN REGRESSOR
knn_regressor = KNeighborsRegressor()
knn_regressor.fit(X_train, y_train)
knn_pred = knn_regressor.predict(X_test)

#EXTRATRESS REGRESSOR
ext_regressor = ExtraTreesRegressor()
ext_regressor.fit(X_train, y_train)
ext_pred = ext_regressor.predict(X_test)

#ADABOOST REGRESSOR (BASE ESTIMATOR = DECISION TREE REGRESSOR)
adb_regressor = AdaBoostRegressor(DecisionTreeRegressor())
adb_regressor.fit(X_train, y_train)
adb_pred = adb_regressor.predict(X_test)

#ADABOOST REGRESSOR (BASE ESTIMATOR = RANDOM FOREST REGRESSOR)
adb_regressor2 = AdaBoostRegressor(RandomForestRegressor())
adb_regressor2.fit(X_train, y_train)
adb_pred2 = adb_regressor2.predict(X_test)

#ADABOOST REGRESSOR (BASE ESTIMATOR = KNN REGRESSOR)
adb_regressor3 = AdaBoostRegressor(KNeighborsRegressor())
adb_regressor3.fit(X_train, y_train)
adb_pred3 = adb_regressor3.predict(X_test)

#ADABOOST REGRESSOR (BASE ESTIMATOR = EXTRA TREES REGRESSOR)
adb_regressor4 = AdaBoostRegressor(ExtraTreesRegressor())
adb_regressor4.fit(X_train, y_train)
adb_pred4 = adb_regressor4.predict(X_test)
```

## 3.2. Model Evaluation

```python
dct = model_evaluation(y_test, dct_pred, "Decision Tree Regressor")
rf = model_evaluation(y_test, rf_pred, "Random Forest Regressor")
knn = model_evaluation(y_test, knn_pred, "KNN Regressor")
ext = model_evaluation(y_test, ext_pred, "ExtraTrees Regressor")
adb = model_evaluation(y_test, adb_pred, "AdaBoost Regressor (Baseline = Decision Tree Regressor)")
adb2 = model_evaluation(y_test, adb_pred2, "AdaBoost Regressor (Baseline = Random Forest Regressor)")
adb3 = model_evaluation(y_test, adb_pred3, "AdaBoost Regressor (Baseline = KNN Regressor)")
adb4 = model_evaluation(y_test, adb_pred4, "AdaBoost Regressor (Baseline = Extra Trees Regressor)")
```

```python
evaluation =dct.append(rf).append(knn).append(ext).append(adb).append(adb2).append(adb3).append(adb4).sort_values(["RMSE"], ascending = True).reset_index().drop(columns = "index")
evaluation
```

We can see that the AdaBoost model with the Base Model Decision Trees has the lowest RMSE value of 0.024954. So next time I will try to do hyperparameter tuning to get the best parameters from the model, because previously I only used default parameters.

## 3.3. Hyperparameter Tunning

```python
adb_regressor.get_params()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from pprint import pprint
import numpy as np

base_estimator = [DecisionTreeRegressor(), ExtraTreesRegressor(), RandomForestRegressor, KNeighborsRegressor()]

# learning rate 
learning_rate = [float(x) for x in np.linspace(start = 0, stop = 1, num = 100)]

# Loss function
loss = ['linear', 'square', 'exponential']


# Membuat random grid
random_grid = {'base_estimator':base_estimator, 
               'learning_rate':learning_rate, 
               'loss':loss}
pprint(random_grid)
```

```python
# Gunakan random grid untuk mencari parameter terbaik
# Parameter dari random search, menggunakan 3 fold cross validation, 
# cari di 100 kombinasi berbeda, dan gunakan semua inti yang tersedia
adb_random = RandomizedSearchCV(estimator = adb_regressor, param_distributions = random_grid, 
                               n_iter = 100, cv = 3, verbose=2, 
                               n_jobs = -1)

# Fit the random search model
adb_random.fit(X_train, y_train)
```

```python
adb_best_params = adb_random.best_params_
adb_best_params
```

```python
adb_regressor_tunning = AdaBoostRegressor(base_estimator = adb_best_params.get('base_estimator'),
                                         loss = adb_best_params.get('loss'),
                                         learning_rate = adb_best_params.get('learning_rate'))
```

```python
adb_regressor_tunning.fit(X_train, y_train)
```

```python
adb_tunning_pred = adb_regressor_tunning.predict(X_test)
```

```python
adb_tunning = model_evaluation(y_test, adb_tunning_pred, "Tunning_AdaBoost Regressor (baseline = Decision Tree Regressor)")
```

```python
evaluation.append(adb_tunning).sort_values(["RMSE"], ascending = True).reset_index().drop(columns = "index")
```

```python
adb_regressor_tunning.fit(X_scaled, y)
```
